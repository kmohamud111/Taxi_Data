```python
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
```

## Load the Data

```python
Cab_Data = pd.read_csv('Cab_Data.csv')
Transaction_ID = pd.read_csv('Transaction_ID.csv')
city = pd.read_csv('City.csv')
Customer_ID = pd.read_csv('Customer_ID.csv')

```

#  Create hypotheses

1.  Effect of Gender on Payment Mode Preference


2.  Age and Income Influence on Cab Company Preference
  

3.  Trip Price Variation in Different Cities
   

4.  Population Density and Cab Usage
  

5.  Trip Distance vs. Price Charged
 

## Merge the Dat

```python
merged1 = pd.merge(Transaction_ID, Cab_Data, on='Transaction ID')
merged2 = pd.merge(merged1, Customer_ID, on='Customer ID')
df = pd.merge(merged2, city, on='City')
```

```python
# Card vs Cash base on Gender

card_vs_cash = df.groupby(['Gender', 'Payment_Mode'])['Transaction ID'].count()
card_vs_cash.plot(kind='bar', color = 'r')
plt.title('Card vs Cash')
plt.xlabel('Type of Payment')
plt.ylabel('Count')
```

```python
# Median Age on each Company

df.groupby(['Company'])['Age'].median().astype(int)
```

```python
#profile on each Company / by 1000 for easy visualization

company_profit = df.groupby('Company')[['Price Charged']].sum()/1000

company_profit.plot(kind='bar', color = 'b')
plt.title('Revenue per Company')
plt.xlabel('Company')
plt.ylabel('Price Charged')
```

```python
city_usage = df.groupby(['Population', 'City'])['Transaction ID'].count().reset_index().sort_values(by='Population', ascending=False)
plt.bar(city_usage.index, city_usage['Transaction ID'], color='b', label='Transaction Count')
plt.xticks(city_usage.index, city_usage['City'], rotation=45, ha='right')
plt.xlabel('City')
plt.ylabel('Transaction Count')
plt.title('Transaction Count by City')
plt.legend()
```
